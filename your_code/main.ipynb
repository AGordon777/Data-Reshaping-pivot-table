#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
# your answer here - crude method to amend the dataset first
def find_max(row):
    if row['Salary'] == employee.pivot_table(index= ['Department'], values = ['Salary'], aggfunc = {'Salary':'max'}).loc[row['Department']]['Salary']:
        return 'Maximum'
    else:
        return 'Not'
```

```python
employee.pivot_table(index= ['Department'], values = ['Salary'], aggfunc = {'Salary':'max'}).loc[['HR']]['Salary']
```

```python
employee['Maximum?'] = employee.apply(find_max, axis=1)
employee
```

```python
not_max = employee[employee['Maximum?']=='Not']
not_max
```

```python
not_max.pivot_table(index= ['Department'], values = ['Salary']).round(2)
```

Alternative: Domenic's code - way more elegant!!!
I forgot you can apply a function within a pivot table, so that it is only applied to an already grouped dataset

```python
def mean_noMax(col):
    col = col[col != col.max()].mean()
    return round(col,2)
    
employee.pivot_table(index=['Department'], aggfunc = {'Salary':mean_noMax}).reset_index()
```
